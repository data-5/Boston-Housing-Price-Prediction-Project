Robust Regression

```python
pd.options.display.float_format = '{:,.2f}'.format
df.head()
```

```python
X= df['RM'].values.reshape(-1,1)
y=df['MEDV'].values
```

```python
from sklearn.linear_model import RANSACRegressor
```

```python
ransac = RANSACRegressor()
```

```python
ransac.fit(X,y)
```

```python
inlier_mask = ransac.inlier_mask_
outlier_mask =np.logical_not(inlier_mask)
```

```python
line_X= np.arange(3,10,1)
line_y_ransac = ransac.predict(line_X.reshape(-1,1))
```

```python
sns.set(style='darkgrid',context ='notebook')
plt.figure(figsize=(12,10))
plt.scatter(X[inlier_mask],y[inlier_mask],c='blue',marker='o',label='inliers')
plt.scatter(X[outlier_mask],y[outlier_mask],c='brown',marker='s',label='outliers')
plt.plot(line_X ,line_y_ransac,color='red')
plt.xlabel('average number of rooms per dwelling')
plt.ylabel("Medium value of owner-occupied home is $1000's")
plt.legend(loc='upper left')
plt.show()
```

```python
ransac.estimator_.coef_
```

```python
ransac.estimator_.intercept_
```

Performance Evaluation of Regression Model

```python
from sklearn.model_selection import train_test_split
```

```python
# X = df['LSTAT'].values.reshape(-1,1)
X = df.iloc[:,:-1].values
```

```python
y = df['MEDV'].values
```

```python
X_train,X_test,y_train,y_test =train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_train_prod = lr.predict(X_train)
```

```python
y_test_prod =lr.predict(X_test)
```

Method 1: Residual Analysis

```python
plt.figure(figsize=(12,8))
plt.scatter(y_train_prod,y_train_prod - y_train, c='blue',marker='o', label='Training data')
plt.scatter(y_test_prod,y_test_prod -y_test, c='orange',marker ='*', label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0,xmin=10,xmax=50,lw=2,color='k')
plt.xlim([-10,50])
plt.show()
```

Method 2: Mean Squared Error(MSE)

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_train,y_train_prod)
```

```python
mean_squared_error(y_test,y_test_prod)
```

Method 3: Coefficient of Determination

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_train,y_train_prod)
```

```python
r2_score(y_test,y_test_prod)
```

What does a Near Perfect Model Looks like?

```python
generate_random = np.random.RandomState(0)
x = 10 * generate_random.rand(1000)
y = 3* x+ np.random.randn(1000)
plt.figure(figsize=(10,8))
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept =True)
model.fit(x_train.reshape(-1,1),y_train)

y_train_prod = model.predict(x_train.reshape(-1,1))
y_test_prod = model.predict(x_test.reshape(-1,1))
```

```python
plt.figure(figsize =(12,8))
plt.scatter(y_train_prod,y_train_prod - y_train,c='blue', marker='o',label='Training data')
plt.scatter(y_test_prod,y_test_prod-y_test,c='orange',marker='*',label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc ='upper left')
plt.hlines(y=0,xmin=3,xmax=33,lw=2,color='k')
plt.xlim([-5,35])
plt.ylim([-25,15])
plt.show()
```

Method 2: Mean Squared Error(MSE)

```python
mean_squared_error(y_train,y_train_prod)
```

```python
mean_squared_error(y_test,y_test_prod)
```

Method 3: Coefficient of Determination

```python
r2_score(y_train,y_train_prod)
```

```python
r2_score(y_test,y_test_prod)
```
